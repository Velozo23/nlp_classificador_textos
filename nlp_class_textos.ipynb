```python
import pandas as pd
```

```python
perguntas = pd.read_csv('stackoverflow_perguntas.csv')
```

```python
perguntas.sample(10)
```

```python
print(len(perguntas))
```

```python
#Conta quantas tags diferentes existem na coluna Tags
print(len(perguntas.Tags.unique()))
```

```python
#Verifica tags unicas utilizando a funcao unique
lista_de_tags = list()

for tags in perguntas.Tags.unique():
    for tag in tags.split():
        if tag not in lista_de_tags:
            lista_de_tags.append(tag)

print(lista_de_tags)
```

```python
#Classificando a tag node_js e criando uma coluna para essa tag

node_js = list()

for linha_tag in perguntas.Tags:
    if "node.js" in linha_tag:
        node_js.append(1)
    else:
        node_js.append(0)

perguntas["node_js"] = node_js

perguntas
```

```python
#Verifica se a tag
def nova_coluna(lista_tags, dataframe, nome_tags):
    for tag in lista_tags:
        coluna = list()
        for linha_tag in dataframe[nome_tags]:
            if tag in linha_tag:
                coluna.append(1)
            else:
                coluna.append(0)

        dataframe[tag] = coluna 
nova_coluna(lista_de_tags, perguntas, "Tags")
perguntas.sample(10)
```
